#### Imports the necessary modules and saves the imported table as  a dataframe

```python
import io
import requests
import pandas as pd # Use pandas to import table
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header = 0, flavor=['lxml'] )[0]

df.head(15)
```

#### Need to Clean up the Dataframe as instructed in several steps

```python

# Renames the columns per Instructions
df.rename(columns={'Postcode':'PostalCode','Neighbourhood':'Neighborhood'}, inplace = True)

# Drop rows that have Not Assigned in the Borough Columns

df = df[df.Borough != 'Not assigned']
df.reset_index(drop = True, inplace = True)


# All Neighborhoods that are Not Assigned should be updated with Borough

for ct in range(0, df.shape[0]):
    if (df.loc[ct,'Neighborhood'] == 'Not assigned'):
        df.replace(df.loc[ct,'Neighborhood'], df.loc[ct,'Borough'],inplace=True)
```

```python
# Combines rows into a single row separated by a comma 

df['Neighborhood'] = df[['PostalCode','Borough','Neighborhood']].groupby(['PostalCode','Borough'])['Neighborhood'].transform(lambda x: ','.join(x)) 
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

df.head(20)
```
